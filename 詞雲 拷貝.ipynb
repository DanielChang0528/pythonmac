### 顯示中文詞雲

```python
with open("/Users/danielchang/Library/Mobile Documents/com~apple~TextEdit/Documents/1-2.txt") as fp:
    text=fp.read()
```

```python
text      #文字檔內容
```

```python
cut_text=' '.join(jieba.cut(text)) #分割字串
```

```python
wd_2=WordCloud(font_path="/Library/Fonts/Songti.ttc",width=800,height=600).generate(cut_text)
```

```python
wd_2_2=wd_2.to_image()
```

```python
plt.imshow(wd_2_2)
```

### 改變背景顏色

```python
wd_3=WordCloud(font_path="/Library/Fonts/Songti.ttc",
               background_color='blue'
               ,width=800,height=600).generate(cut_text) #加入background_color
```

```python
wd_3_3=wd_3.to_image()
```

```python
plt.imshow(wd_3_3)
```

### 去除座標軸

```python
wd_3=WordCloud(font_path="/Library/Fonts/Songti.ttc",
               background_color='blue'
               ,width=800,height=600).generate(cut_text)
```

```python
wd_3_3=wd_3.to_image()
```

```python
plt.axis("off")  ###去除座標軸
plt.imshow(wd_3_3)
```
